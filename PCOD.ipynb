```python
colors = ['#670067','#008080']
```

```python
def bar_plot(variable):
    """
     input: variable example : Target
     output: bar plot & value count

    """
    # Get feature
    var = data[variable]
    # Count number of categorical variable(value/sample)
    varValue = var.value_counts()
    # Visualize
    plt.figure(figsize = (9,3))
    plt.bar(varValue.index,varValue,color=colors)
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel("Count")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
category = ["Target", "Pregnant(Y/N)", "Weight gain(Y/N)", "hair growth(Y/N)", "Skin darkening (Y/N)", "Hair loss(Y/N)", "Pimples(Y/N)", "Fast food (Y/N)", "Reg.Exercise(Y/N)", "Blood Group"]
for c in category:
    bar_plot(c)
```

```python
def plot_hist(variable):
    plt.figure(figsize = (9,3))
    plt.hist(data[variable], bins = 50,color=colors[1])
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distribution with hist".format(variable))
    plt.show()
```

```python
numericVar = ["Age (yrs)", "Weight (Kg)","Marraige Status (Yrs)"]
for n in numericVar:
    plot_hist(n)
```

```python
from sklearn.feature_selection import SelectKBest, chi2

chi_selector = SelectKBest(score_func=chi2, k=10)
x_chi = chi_selector.fit_transform(x, y)

chi_scores = pd.DataFrame({
    "Feature": x.columns,
    "ChiSquare Score": chi_selector.scores_
})

chi_scores = chi_scores.sort_values(by="ChiSquare Score", ascending=False)
print(chi_scores)
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
random_state = 42
classifier = [DecisionTreeClassifier(random_state = random_state),
             SVC(random_state = random_state),
             RandomForestClassifier(random_state = random_state),
             LogisticRegression(random_state = random_state),
             KNeighborsClassifier()]
```

```python
dt= {"min_samples_split" : range(10,500,20),"max_depth": range(1,20,2)}
```

```python
svc = {"kernel" : ["rbf"],"gamma": [0.001, 0.01, 0.1, 1],"C": [1,10,50,100,200,300,1000]}
```

```python
rf= {"max_features": ['auto', 'sqrt', 'log2'],"n_estimators":[300,500],"criterion":["gini"],'max_depth' : [4,5,6,7,8,9,10,12],}
```

```python
logreg= {"C":np.logspace(-3,3,7),"penalty": ["l1","l2"]}
```

```python
knn= {"n_neighbors": np.linspace(1,19,10, dtype = int).tolist(),"weights": ["uniform","distance"],"metric":["euclidean","manhattan"]}
```

```python
classifier_param = [dt,svc, rf,logreg,knn]
```

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV
cv_result = []
best_estimators = []
for i in range(len(classifier)):
    clf = GridSearchCV(classifier[i], param_grid=classifier_param[i], cv = StratifiedKFold(n_splits = 10),
                       scoring = "accuracy", n_jobs = -1,verbose = 1)
    clf.fit(x_train,y_train)
    cv_result.append(round(clf.best_score_*100,2))
    best_estimators.append(clf.best_estimator_)
    print(cv_result[i])
```

```python
best_estimators
```

```python
dt = best_estimators[0]
svm = best_estimators[1]
rf = best_estimators[2]
logreg = best_estimators[3]
knn = best_estimators[4]
```

```python
import xgboost
xgb_clf = xgboost.XGBRFClassifier(max_depth=3, random_state=random_state)
xgb_clf.fit(x_train,y_train)
acc_xgb_clf_train = round(xgb_clf.score(x_train, y_train)*100,2)
acc_xgb_clf_test = round(xgb_clf.score(x_test,y_test)*100,2)
cv_result.append(acc_xgb_clf_train)
print("Training Accuracy: % {}".format(acc_xgb_clf_train))
print("Testing Accuracy: % {}".format(acc_xgb_clf_test))
```
